```python
from Bio import Entrez 
from Bio import Medline
#imports functions which mine pubmed data 
```

```python
query_1 = input("Keyword 1: ")
query_2 = input("Keyword 2: ")

def combined_query(query_1, query_2):
    return query_1  + " " + query_2
 
print(combined_query(query_1, query_2))
```

```python
def search(combined_query):
    Entrez.email = 'adamsorbie@tum.de'
    handle = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax='100',
                            retmode='xml', 
                            term=combined_query)
    results = Entrez.read(handle)
    return results
# search 
```

```python
def pubmed_fetch(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'adamsorbie@tum.de'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results
```

```python
import json 
results = search(combined_query)
id_list = results['IdList']
papers = pubmed_fetch(id_list)
print(json.dumps(papers, indent=4, separators=(',',':')))
with open(, 'w') as outfile:
    json.dump(papers, outfile)
#writes fetched results to JSON file 
```
